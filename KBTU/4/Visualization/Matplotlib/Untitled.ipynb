```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,1, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r--')
plt.title('Test')
plt.xlabel('X coord')
plt.ylabel("Y coord");
plt.show() # %matplotlib inline для этого нужен
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-*')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'r--');
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0,0.8, 0.8])
axes.plot(x,y,'b')
axes.set_xlabel("X label")
axes.set_ylabel('Y label')
axes.set_title('MATPLOT');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Axes1 x label')
axes1.set_ylabel('Axes1 y label')
axes1.set_title('Axes1')

axes2.plot(x, y, 'r--')
axes2.set_xlabel('Axes2 x label')
axes2.set_ylabel('Axes2 y label')
axes2.set_title('Axes2');
```

```python
fix, axes = plt.subplots()
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title');
```

```python
# fix, axes = plt.subplots(nrows=1, ncols=2)
# Empty canvas of 1 by 2 subplots
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'r--')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title("Title");
    
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y, 'b-+')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title("Title");
    
fig
plt.tight_layout();
```
