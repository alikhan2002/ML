```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
# iris.index.names
```

```python
g = sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris, palette='viridis', hue='species')
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
# iris.head()
```
